```python
import pandas as pd
from sqlalchemy import create_engine

# Conexión a la base de datos de Docker
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')

# Leer Parquet y enviarlo a SQL
# Nota: Si te da error aquí, ejecuta en una celda: !pip install pyarrow
df_green = pd.read_parquet('green_tripdata_2025-11.parquet')
df_green.to_sql(name='green_tripdata', con=engine, if_exists='replace', index=False)

#cargar zonas, Si el archivo está en la misma carpeta que el notebook:
df_zones = pd.read_csv('taxi_zone_lookup.csv')
df_zones.to_sql(name='zones', con=engine, if_exists='replace', index=False)

# Para verificar que se cargó bien, mirar las primeras 5 filas:
# df_zones.head()

print("¡Carga completada con éxito!")
```
